```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("movies.csv")
df
```

```python
# Let'slook at the data (fist 5 lines)
df.head()
```

```python
## Let'slook at the data (last 5 lines)
df.tail()
```

```python
#Let's see any missing data

for col in df.columns:
    pct_missing =  np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#Let's check colum names

df.columns
```

```python
#Let's check count of columns
len(df.columns)
```

```python
#Let's check NaN

df.isnull().sum()
```

```python
#Let's fill NaN

df["rating"] = df["rating"].fillna(value="0")
df["released"] =df["released"].fillna(value="01.01.1900")
df["score"] = df["score"].fillna(value="0")
df["votes"] = df["votes"].fillna(value="0")
df["writer"] = df["writer"].fillna(value="no input")
df["star"] = df["star"].fillna(value="no input")
df["country"] = df["country"].fillna(value="no input")
df["budget"] = df["budget"].fillna(value="0")
df["gross"] = df["gross"].fillna(value="0")
df["company"] = df["company"].fillna(value="no input")
df["runtime"] = df["runtime"].fillna(value="0")
```

```python
#We have to be sure data has no NaN

df.isnull().sum()
```

```python
#Data types of columns
df.dtypes
```

```python
#change datatype of columns

df["budget"] = df["budget"].astype("int64")
df["gross"] = df["gross"].astype("int64")
df.dtypes
```

```python
df.dtypes
```

```python
#Create correct year column

df[['month_day','year']] = df.released.str.split(",",expand=True,)
df["year_correct"] = df["year"].astype(str).str[:5]
df.head()
```

```python
#Let's sort gross column as ascending values

df.sort_values(by=["gross"], inplace=False , ascending = False)
```

```python
pd.set_option("display.max_rows",None)
```

```python
#Drop any duplicates
df["company"].drop_duplicates().sort_values(ascending=False)
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df["budget"],y=df["gross"],color="r")
plt.title("Budget vs Gross Earnings")
plt.xlabel("Budget Earnings")
plt.ylabel("Gross Earnings")
plt.show
```

```python
#Plot budget vs gross using seaborn

sns.regplot(x="budget",y="gross" ,data=df,scatter_kws = {"color": "red"}, line_kws={"color":"blue"})
```
